```python
import pandas_geojson as dj
import json

geojson = dj.read_geojson('/home/jeffin/Documents/ICFOSS/MalayalamDialectMap/src/assets/panchayath.geojson')
df = geojson.to_dataframe()

# Extract data from columns
lsgd_data = df['properties.LSGD'].tolist()
district_data = df['properties.DISTRICT'].tolist()
taluk_data = df['properties.BLOCK'].tolist()

# Combine data into nested dictionary
nested_data = {'PANCHAYATH': []}
for i, lsgd in enumerate(lsgd_data):
  nested_data['PANCHAYATH'].append({'NAME':lsgd_data[i],'DISTRICT': district_data[i],"BLOCK":taluk_data[i]})

# Print the nested dictionary
print(nested_data)

# Save as JSON file
with open('nested_lsgd_district_data.json', 'w') as outfile:
  json.dump(nested_data, outfile, indent=4)  # Add indentation for readability

print("Nested LSGD and DISTRICT data saved to nested_lsgd_district_data.json")
```
